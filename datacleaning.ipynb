```python
#import library
import pandas as pd
```

```python
#load csv file
df = pd.read_csv(r"C:\Users\SK\Documents\superstore_sales.csv")

#display first 5 rows to check that file loaded correctly
df.head(5)
```

```python
#check summary of structure of dataframe and data types 
df.info()
```

```python
#convert data type of columns containing data from object to datatime
df['Order Date']= pd.to_datetime(df['Order Date'],format ='%d/%m/%Y')
df['Ship Date']= pd.to_datetime(df['Ship Date'],format ='%d/%m/%Y')

#check changed data type 
df.dtypes
```

```python
#count missing values in column Postal Code
count=df['Postal Code'].isnull().sum()
count
```

```python
#find the number of duplicates
df.duplicated().sum()
```

```python
#handling missing values
df.dropna(subset=['Postal Code'], inplace=True)
```

```python
#check final summary of structure of dataframe and data types
df.info()
```
